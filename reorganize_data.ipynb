```python
import pandas as pd
import numpy as np
```

```python
data_1 = pd.read_csv(r'.\TrafficLabelling\Friday-WorkingHours-Afternoon-DDos.pcap_ISCX.csv', encoding="latin1")
data_2 = pd.read_csv(r'.\TrafficLabelling\Friday-WorkingHours-Afternoon-PortScan.pcap_ISCX.csv', encoding="latin1")
data_3 = pd.read_csv(r'.\TrafficLabelling\Friday-WorkingHours-Morning.pcap_ISCX.csv', encoding="latin1")
data_4 = pd.read_csv(r'.\TrafficLabelling\Monday-WorkingHours.pcap_ISCX.csv', encoding="latin1")
data_5 = pd.read_csv(r'.\TrafficLabelling\Thursday-WorkingHours-Afternoon-Infilteration.pcap_ISCX.csv', encoding="latin1")
data_6 = pd.read_csv(r'.\TrafficLabelling\Thursday-WorkingHours-Morning-WebAttacks.pcap_ISCX.csv', encoding="latin1")
data_7 = pd.read_csv(r'.\TrafficLabelling\Tuesday-WorkingHours.pcap_ISCX.csv', encoding="latin1")
data_8 = pd.read_csv(r'.\TrafficLabelling\Wednesday-WorkingHours.pcap_ISCX.csv', encoding="latin1")
```

```python
data_1.columns
```

```python
for df in [data_1, data_2, data_3, data_4,data_5, data_6, data_7, data_8]:
    df.columns = (
        df.columns
        .str.strip()
        .str.replace(r"\s+", " ", regex=True)
    )
```

```python
print(data_1["Label"].value_counts())
print(data_2["Label"].value_counts())
print(data_3["Label"].value_counts())
print(data_4["Label"].value_counts())
print(data_5["Label"].value_counts())
print(data_6["Label"].value_counts())
print(data_7["Label"].value_counts())
print(data_8["Label"].value_counts())
```

```python
mapping = {
    "BENIGN": "BENIGN",

    # DOS_DDOS
    "DDoS": "DOS_DDOS",
    "DoS Hulk": "DOS_DDOS",
    "DoS GoldenEye": "DOS_DDOS",
    "DoS slowloris": "DOS_DDOS",
    "DoS Slowhttptest": "DOS_DDOS",

    # PROBE_PORTSCAN
    "PortScan": "PROBE_PORTSCAN",

    # CRED_WEB
    "FTP-Patator": "CRED_WEB",
    "SSH-Patator": "CRED_WEB",
    "Web Attack - Brute Force": "CRED_WEB",
    "Web Attack - XSS": "CRED_WEB",
    "Web Attack - Sql Injection": "CRED_WEB",

    # R2L_U2R_OTHER
    "Bot": "R2L_U2R_OTHER",
    "Infiltration": "R2L_U2R_OTHER",
    "Heartbleed": "R2L_U2R_OTHER"
}
```

```python
data_1["Label"] = data_1["Label"].map(mapping)
data_2["Label"] = data_2["Label"].map(mapping)
data_3["Label"] = data_3["Label"].map(mapping)
data_4["Label"] = data_4["Label"].map(mapping)
data_5["Label"] = data_5["Label"].map(mapping)
data_6["Label"] = data_6["Label"].map(mapping)
data_7["Label"] = data_7["Label"].map(mapping)
data_8["Label"] = data_8["Label"].map(mapping)
print(data_1["Label"].value_counts())
print(data_2["Label"].value_counts())
print(data_3["Label"].value_counts())
print(data_4["Label"].value_counts())
print(data_5["Label"].value_counts())
print(data_6["Label"].value_counts())
print(data_7["Label"].value_counts())
print(data_8["Label"].value_counts())
```

```python
data = pd.concat([data_1, data_2, data_3, data_4, data_5, data_6, data_7, data_8], axis=0)
data = data.reset_index(drop=True)
data = data.replace([np.inf, -np.inf], np.nan)
data.info()
```

```python
# Tổng số NaN trên toàn DataFrame
print("Tổng số NaN:", data.isna().sum().sum())

# NaN theo từng cột
print("\nNaN theo từng cột:")
print(data.isna().sum()[data.isna().sum() > 0])
```

```python
data[["Flow Bytes/s", "Flow Packets/s"]] = data[["Flow Bytes/s", "Flow Packets/s"]].fillna(0)
```

```python
print(data["Label"].value_counts(dropna=False))
print(data["Label"].value_counts(normalize=True))
```

```python
data=data.dropna(subset=["Label"])
data_binary = data.copy()

# Chuẩn hoá label trước (phòng trường hợp có space/hoa-thường lạc)
data_binary["Label"] = data_binary["Label"].str.strip().str.upper()

# Gom nhóm
data_binary["Label"] = data_binary["Label"].apply(
    lambda x: "BENIGN" if x == "BENIGN" else "ATTACK"
)

print(data_binary["Label"].value_counts())
```

```python
print(data_binary["Label"].value_counts())
print(data["Label"].value_counts())
attack_total = data["Label"].value_counts().drop("BENIGN").sum()
print("Tổng ATTACK:", attack_total)
```

```python
data_binary.to_csv(r'.\TrafficLabelling\data_binary.csv', index=False)
data.to_csv(r'.\TrafficLabelling\data_multi_class.csv', index=False)
```
